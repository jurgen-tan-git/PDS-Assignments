```python
import pandas as pd
```

```python
coursegrad = pd.read_csv(r"graduates-from-university-first-degree-courses-by-type-of-course.csv",
                        na_values = "na",
                        index_col = 0)
coursegrad = coursegrad[coursegrad.year >= 2014].set_index("year")
coursegrad = coursegrad.fillna(0)
coursegrad.no_of_graduates = coursegrad.no_of_graduates.astype("int32")
coursegrad
```
